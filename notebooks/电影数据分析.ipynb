# 2. 电影情况分析

## 2.1 电影时长分布

```python
# 电影时长分布
movie_duration = df2.duration
movie_duration
```

```python
# 创建画布
plt.figure(figsize=(20, 8), dpi=100)
# 设置组距
distance = 5
# 计算组数
group_num = int((max(movie_duration) - min(movie_duration)) / distance)
# 绘制直方图
plt.hist(movie_duration, bins=group_num, range=(min(movie_duration), max(movie_duration)), color='#FF7F50')
# 设置x轴刻度
plt.xticks(np.arange(min(movie_duration), max(movie_duration), distance), rotation=45)
# 设置x轴刻度标签
plt.xlabel('Duration')
# 填加网格
plt.grid(linestyle='--', alpha=0.5)
# 填加标题
plt.title('Movie Duration')
# 保存图片
plt.savefig('movie_duration.png')
# 显示图片
plt.show()
```

## 2.2 电影类型分析

> 原始数据中，每部电影均对应多个题材，需对原始数据进行处理，使之能够每个电影均对应一个题材

```python
genre_data = pd.DataFrame(columns=['genre', 'gross'])
for index, row in df2.iterrows():
    genres = row['genres'].split('|')
    genre_count = len(genres)
    new_line = {}
    new_line['genre'] = genres
    new_line['gross'] = [row['gross']] * genre_count
    genre_data = genre_data.append(pd.DataFrame(new_line), ignore_index=True)
genre_data.to_csv('../output/genre_data.csv', index=None)
```

```python
# 按题材分类，统计数量
genre_count = genre_data.groupby('genre')['genre'].count()
plt.figure(figsize=(15, 10), dpi=80)
genre_count.plot(kind='bar')
plt.xticks(rotation=45)
plt.savefig('genre_count.png')
plt.show()
```

```python
# 按题材分类，统计票房收入
genre_gross = genre_data.groupby('genre')['gross'].sum()
plt.figure(figsize=(15, 10), dpi=80)
genre_gross.plot(kind='bar')
plt.xticks(rotation=45)
plt.savefig('genre_gross.png')
plt.show()
```
